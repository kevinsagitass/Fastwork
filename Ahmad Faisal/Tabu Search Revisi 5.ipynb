```python
# Import Library yang diperlukan

import numpy as np
import pandas as pd
import random as rd
```

# Data Preprocessing

```python
# Data Preprocessing

wilayah = pd.read_excel('Book1.xlsx', sheet_name = "jarak", header = None, usecols = [0, 1, 2], skiprows = [0, 1, 2])

wilayah['gabungan'] = wilayah[0] + ' - ' + wilayah[1] + ' - ' + wilayah[2]

wilayah['gabungan']
```

```python
# Menggabungkan data menjadi 1 DataFrame

df = pd.read_excel('Book1.xlsx', sheet_name = "jarak", header = None, names = wilayah['gabungan'].tolist(),
                       usecols = [i for i in range(3, 264)], skiprows = [0, 1, 2])

df.index = wilayah['gabungan']

df.head()
```

```python
#Pembuatan DataFrame untuk fungsi objektif

#Skor
k = np.sum(df.values * pd.read_excel('Book1.xlsx',
                             sheet_name = "bobot_permintaan",
                             header = None,
                             skiprows = [0, 1, 2],
                             usecols = [i for i in range(1, 262)]).T.values,axis = 0)

df_obj = pd.DataFrame()
df_obj['skor'] = k
df_obj.index = wilayah['gabungan']


#Macet
df_obj['macet'] = pd.read_excel('Book1.xlsx',
                             sheet_name = "bebas_macet",
                             header = None,
                             skiprows = [0, 1, 2],
                             usecols = [i for i in range(1, 262)]).T.values


#Banjir
df_obj['banjir'] = pd.read_excel('Book1.xlsx',
                             sheet_name = "bebas_banjir",
                             header = None,
                             skiprows = [0, 1, 2],
                             usecols = [i for i in range(1, 262)]).T.values

df_obj.head()
```

```python
#Skor suatu wilayah didefinisikan sebagai banjir x 3 + macet x 2 + skor x 1

df_skor = df_obj['skor'] * 1 + df_obj['banjir'] * 3 + df_obj['macet'] * 2

df_skor.sort_values(ascending = False)
```

# Tabu Search

```python
# Membuat 2 angka random lokasi awal
def generate_initial_solution(n_index = len(df_skor)):
    return [rd.randint(0, n_index), rd.randint(0, n_index)]


# Parameter Tabu Search
stopping_criteria = range(0, 200)    
tabu_tenure = 30
jumlah_loop = range(0, 100)

#Variabel nilai maksimal masing-masing loop
loop_solutions_score = []
current_max_loop_number = 0

# Running sekian kali Tabu Search
for k in jumlah_loop:
    rd.seed(k)
    
    #Reset variabel saat mulai running tabu search baru
    current_solution = generate_initial_solution()
    solution_score = df_skor[current_solution[0]] * df_skor[current_solution[1]] #skor perkalian dua lokasi awal
    current_solution_score = solution_score
    current_best_solution = current_solution
    current_best_solution_score = solution_score
    
    tabu_list = []

    current_best_solution_score = current_solution_score
    tabu_list.append(current_solution)

    # Running Tabu Search dengan iterasi ssebanyak stopping _criteria
    for i in stopping_criteria:

        #Reset variabel saat mulai iterasi baru
        new_solution_1 = []
        new_solution_2 = []
        new_solution_3 = []
        new_solution_4 = []

        solutions_list = []
        solutions_list_score = []


        # Mencari nilai skor tetangga 
        if current_solution[0] == 0:
            new_solution_1 = [len(df_skor)-1, current_solution[1]]
        else:
            new_solution_1 = [current_solution[0]-1, current_solution[1]]

        if not new_solution_1 in tabu_list:
            new_solution_1_score = df_skor[new_solution_1[0]] * df_skor[new_solution_1[1]]

        solutions_list.append(new_solution_1)
        solutions_list_score.append(new_solution_1_score)


        if current_solution[0] == len(df_skor)-1:
            new_solution_2 = [0, current_solution[1]]
        else:
            new_solution_2 = [current_solution[0]+1, current_solution[1]]

        if not new_solution_2 in tabu_list:
            new_solution_2_score = df_skor[new_solution_2[0]] * df_skor[new_solution_2[1]]

        solutions_list.append(new_solution_2)
        solutions_list_score.append(new_solution_2_score)


        if current_solution[1] == 0:
            new_solution_3 = [current_solution[0], len(df_skor)-1]
        else:
            new_solution_3 = [current_solution[0], current_solution[1]-1]  

        if not new_solution_3 in tabu_list:
            new_solution_3_score = df_skor[new_solution_3[0]] * df_skor[new_solution_3[1]]

        solutions_list.append(new_solution_3)
        solutions_list_score.append(new_solution_3_score)  


        if current_solution[1] == len(df_skor)-1:
            new_solution_4 = [current_solution[0], 0]
        else:
            new_solution_4 = [current_solution[0], current_solution[1]+1]   

        if not new_solution_4 in tabu_list:
            new_solution_4_score = df_skor[new_solution_4[0]] * df_skor[new_solution_4[1]]

        solutions_list.append(new_solution_4)
        solutions_list_score.append(new_solution_4_score) 

        
        #Print Informasi lokasi yang telah didapatkan
        if i == len(jumlah_loop)-1:
            print("Run ke-" + str(k+1))

            print("Lokasi 1: " + wilayah['gabungan'][current_best_solution[0]], end = " (")
            if df_obj.loc[wilayah['gabungan'][current_best_solution[0]]][1] == 0:
                print("Banjir dan ", end = "")
            else:
                print("Tidak Banjir dan ", end = "")

            if df_obj.loc[wilayah['gabungan'][current_best_solution[0]]][2] == 0:
                print("Macet)")
            else:
                print("Tidak Macet)")

            print("Lokasi 2: " + wilayah['gabungan'][current_best_solution[1]], end = " (")
            if df_obj.loc[wilayah['gabungan'][current_best_solution[1]]][1] == 0:
                print("Banjir dan ", end = "")
            else:
                print("Tidak Banjir dan ", end = "")

            if df_obj.loc[wilayah['gabungan'][current_best_solution[1]]][2] == 0:
                print("Macet)")
            else:
                print("Tidak Macet)")
            
            print("Current Best Solution Score: " + str(current_best_solution_score))
            print("===")
            print("")


        #Update Tabu list and Score
        current_solution_score = max(solutions_list_score)
        current_solution_index = solutions_list_score.index(max(solutions_list_score))

        if current_best_solution_score <= current_solution_score:
            current_best_solution_score = current_solution_score
            current_best_solution = solutions_list[current_solution_index]

        if not current_best_solution in tabu_list:
            tabu_list.append(current_best_solution)

        current_solution = current_best_solution

        #Menghilangkan Elemen dalam Tabu List dengan Tenure
        if i > tabu_tenure:
            tabu_list.pop(0)
            
    loop_solutions_score.append(current_best_solution_score)
    
    if current_best_solution_score >= max(loop_solutions_score):
        current_max_loop_number = k
        
print("====================")
print("Dari seluruh looping yang dilakukan, diperoleh 2 lokasi dengan skor yang paling tinggi yaitu:")
print("Looping ke-" + str(current_max_loop_number+1) + " dengan skor sebesar " + str(max(loop_solutions_score)))
```
