```python
import pandas as pd 
```

```python
wilayah = pd.read_excel('Book1.xlsx',
                        sheet_name=1,
                        header=None,
                        usecols=[0, 1, 2],
                        skiprows=[0, 1, 2])
wilayah['gabungan'] = wilayah[0] + ' - ' + wilayah[1] + ' - ' + wilayah[2]
```

```python
wilayah['gabungan']
```

```python
df = pd.read_excel('Book1.xlsx',
                   sheet_name=1,
                   header=None,
                   names=wilayah['gabungan'].tolist(),
                   usecols=[i for i in range(3, 264)],
                   skiprows=[0, 1, 2])
```

```python
df.index = wilayah['gabungan']
```

```python
df['banjir'] = pd.read_excel('Book1.xlsx',
                             sheet_name=2, header=None,
                             skiprows=[0, 1, 2],
                             usecols=[i for i in range(1, 262)]).T.values
```

```python
df['macet'] = pd.read_excel('Book1.xlsx',
                             sheet_name=3,
                             header=None,
                            skiprows=[0, 1, 2],
                            usecols=[i for i in range(1, 262)]).T.values
```
